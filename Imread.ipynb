<h1><b>Importing Image using Scipy</b></h1>

```python

from scipy import misc
import imageio
import matplotlib.pyplot as plt

face = misc.face()
imageio.imsave('bee.webp', face)

plt.imshow(face)
plt.show()
```

```python
face.shape
```

```python
face.ndim
```

```python
face.size
```

```python
#now i will try to change its size and color
from scipy import misc, ndimage
import numpy as np

resize_face = ndimage.zoom(face, (0.5, 0.5, 1))
gray_face = np.dot(resize_face[...,:3], [0.2989, 0.5870, 0.1140])

imageio.imsave('bee_resized_grayscale.webp', gray_face)

plt.imshow(gray_face, cmap=plt.get_cmap('gray'))
plt.show()
```
